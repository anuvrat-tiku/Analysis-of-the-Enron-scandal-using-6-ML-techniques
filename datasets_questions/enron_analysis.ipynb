```python
import pickle
import pandas as pd
enron_data = pickle.load(open("../final_project/final_project_dataset.pkl", "r"))
```

```python
# Getting the length of the dictionary = 146. This is the number of data points in the dataset.
print len(enron_data.keys())
print len(enron_data)
```

```python
# We can use Pandas to get the structure of the data and understand its features.
# Convert the data into a data frame
df = pd.DataFrame(enron_data)
df.head()
```

```python
# Pandas shape will give us the dimensions of the data
print "Data Frame dimensions are :", df.shape
# Index will give us the all the rows/features of the data
print "Features of the dataframe are:"
for x in df.index:
    print x

# To get all the columns, simply use the list function in Pandas
print "All the", len(list(df)), "columns/people in the data set:"
list(df)
```

```python
""" Finding POI or Persons of Intrest. The key in the dictionary is the name of the person and the values are a JSON with 
    one of the keys of the JSON being 'poi'.
"""
print "POI classification for all the people in the dataset"
for x in enron_data:
    print x, enron_data[x]['poi']
```

```python
# Counting the number of True POI's
count = 0
for x in enron_data:
    if enron_data[x]['poi'] == True:
        count += 1
print "Number of POI's:", count
```

```python
# Importing the list from USA today.
enron_all_txt = open("../final_project/poi_names.txt", "r")
print "List of all people mentioned in USA today article : http://usatoday30.usatoday.com/money/industries/energy/2005-12-28-enron-participants_x.htm"
for line in enron_all_txt:
    print line.split(" ")
    
```

```python
# Finding which of the people have a quantifiable salary and email address
no_salary_count = 0
for x in enron_data:
    if enron_data[x]['salary'] == 'NaN':
        no_salary_count+=1
        
print "Salary is missing for", no_salary_count, "people."

# Finding which of the people have a quantifiable salary and email address
no_email_count = 0
for x in enron_data:
    if enron_data[x]['email_address'] == 'NaN':
        no_email_count+=1
        
print "Email is missing for", no_email_count, "people."
```

```python
# Finding out how many people have missing total_payments
count_tp = 0
for x in enron_data:
    if enron_data[x]['total_payments'] == 'NaN':
        count_tp += 1
print "People mssing total payments:", count_tp
print "This is", (count_tp*100/len(enron_data)), "% of the total population in the sample dataset."
```

```python
# Finding out how many people have missing total_payments and are POIs
count_tp_poi = 0
for x in enron_data:
    if enron_data[x]['total_payments'] == 'NaN' and enron_data[x]['poi'] == True:
        count_tp_poi += 1
print "People mssing total payments:", count_tp_poi
print "This is", (count_tp_poi*100/len(enron_data)), "% of the total population in the sample dataset who are POIs."
```
